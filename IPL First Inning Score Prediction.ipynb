```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ipl = pd.read_csv('ipl_data.csv')
```

```python
ipl.head()
```

```python
col_to_remove = ['mid', 'venue', 'batsman', 'bowler', 'striker', 'non-striker']
ipl.drop(col_to_remove, axis=1, inplace=True)
ipl.head()
```

```python
ipl['bat_team'].unique()
```

```python
consistent_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians', 'Kings XI Punjab',
       'Royal Challengers Bangalore', 'Delhi Daredevils', 'Sunrisers Hyderabad']
ipl = ipl[(ipl['bat_team'].isin(consistent_teams)) & (ipl['bowl_team'].isin(consistent_teams))]
ipl.head()
```

```python
ipl = ipl[ipl['overs'] >= 5.0]
```

```python
ipl.head()
```

```python
print(ipl['bat_team'].unique())
print(ipl['bowl_team'].unique())
```

```python
from datetime import datetime
ipl['date'] = ipl['date'].apply(lambda x : datetime.strptime(x, '%Y-%m-%d'))
```

```python
ipl.head()
```

```python
encoded_ipl = pd.get_dummies(data=ipl, columns=['bat_team', 'bowl_team'])
```

```python
encoded_ipl.head()
```

```python
encoded_ipl.columns
```

```python
encoded_ipl = encoded_ipl[['date', 'bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils',
       'bat_team_Kings XI Punjab', 'bat_team_Kolkata Knight Riders',
       'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils',
       'bowl_team_Kings XI Punjab', 'bowl_team_Kolkata Knight Riders',
       'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
       'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad', 'runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5',
       'total']]
encoded_ipl.head()
```

```python
x_train = encoded_ipl.drop(['total'], axis=1)[encoded_ipl['date'].dt.year <= 2016]
x_test = encoded_ipl.drop(['total'], axis=1)[encoded_ipl['date'].dt.year >= 2017]
print(x_train.shape)
print(x_test.shape)
```

```python
y_train = encoded_ipl[encoded_ipl['date'].dt.year <= 2016]['total'].values
y_test = encoded_ipl[encoded_ipl['date'].dt.year >= 2017]['total'].values
print(y_train.shape)
print(y_test.shape)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
# Removing the 'date' column
x_train.drop(['date'], axis=1, inplace=True)
x_test.drop(['date'], axis=1, inplace=True)
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()
parameters = {'alpha':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 35, 40]}
ridge_regressor = GridSearchCV(ridge, parameters, scoring='neg_mean_squared_error', cv=5)
ridge_regressor.fit(x_train, y_train)
```

```python
print(ridge_regressor.best_score_)
```

```python
prediction = ridge_regressor.predict(x_test)
print(prediction)
```

```python
sns.distplot(y_test - prediction)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print('MAE : ', mean_absolute_error(y_test, prediction))
print('MSE : ', mean_squared_error(y_test, prediction))
print('RMSE : ', np.sqrt(mean_squared_error(y_test, prediction)))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
parameters = {'alpha':[1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20, 30, 35, 40]}
lasso_regressor = GridSearchCV(lasso, parameters, scoring='neg_mean_squared_error', cv=5)

lasso_regressor.fit(x_train,y_train)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
prediction_ls = lasso_regressor.predict(x_test)
prediction_ls
```

```python
sns.distplot(y_test - prediction_ls)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print('MAE : ', mean_absolute_error(y_test, prediction_ls))
print('MSE : ', mean_squared_error(y_test, prediction_ls))
print('RMSE : ', np.sqrt(mean_squared_error(y_test, prediction_ls)))
```

```python
import joblib
joblib.dump(lasso_regressor, 'ipl_first_inning_score_predictor.pkl')
```
